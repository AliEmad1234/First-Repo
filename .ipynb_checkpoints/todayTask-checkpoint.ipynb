```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

# Using Linear SVM

```python
dataset = pd.read_csv('todayTask/heart.csv')
dataset.corr()
```

```python
features = dataset.iloc[:,0:-1].values
goal = dataset.iloc[:,-1].values

# standrize the values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features = scaler.fit_transform(features)
```

```python
#import seaborn as sns
#df = pd.DataFrame({'Age':features[:,2] , 'Goal':features[:,5]})
#sns.lmplot(x='Age', y='Goal', data=df,
#            fit_reg=False );   
```

```python
# split into train and test
from sklearn.model_selection import train_test_split
train_set, test_set, goal_train, goal_test = train_test_split(features,goal,test_size =0.25,random_state=0)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel ='linear', random_state = 0)
classifier.fit(train_set, goal_train)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
for K in [1, 3, 5, 10, 20, 100, 150, len(train_set)]:
    classifier = KNeighborsClassifier(n_neighbors=K)
    classifier.fit(train_set, goal_train)
    cm = confusion_matrix(goal_test, classifier.predict(test_set))
    print('K = %d\n-------------'%K)
    print(pd.DataFrame(cm))
    print("model accuracy : {:.4f}".format(classifier.score(test_set, goal_test)))
    print('-------------\n\n')
```

# Using Logistic Reg

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(random_state = 0)
logistic.fit(train_set, goal_train)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, logistic.predict(test_set))
pd.DataFrame(cm)
```

```python
#%%time
logistic.score(test_set, goal_test)
```

# Bank

```python
bank = pd.read_csv('todayTask/bank-full.csv' , sep=';')

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
bank.iloc[:,-1] = encoder.fit_transform(bank.iloc[:,-1])
bank.iloc[:,2] = encoder.fit_transform(bank.iloc[:,2])
bank.iloc[:,3] = encoder.fit_transform(bank.iloc[:,3])

bank.corr()
```

```python
features = bank.iloc[:,[0,2,3,5]].values
goal = bank.iloc[:,-1].values

# standrize the values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features = scaler.fit_transform(features)
```

```python
# split into train and test
from sklearn.model_selection import train_test_split
train_set, test_set, goal_train, goal_test = train_test_split(features,goal,test_size =0.20,random_state=0)
```

# Using Linear SVM

```python
from sklearn.svm import SVC
classifier = SVC(kernel ='linear', random_state = 0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
for K in [1, 3, 5, 10, 20, 100, 150, len(test_set)]:
    classifier = KNeighborsClassifier(n_neighbors=K)
    classifier.fit(train_set, goal_train)
    print("model accuracy : {:.4f}".format(classifier.score(test_set, goal_test)))

    cm = confusion_matrix(goal_test, classifier.predict(test_set))
    print('K = %d\n-------------'%K)
    print(pd.DataFrame(cm))
    print('-------------\n\n')
```

# Using Logistic

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(random_state = 0)
logistic.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, logistic.predict(test_set))
pd.DataFrame(cm)

logistic.score(test_set, goal_test)
```

# Using Naive Bays

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state=0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using Gaussian SVM

```python
from sklearn.svm import SVC
classifier = SVC(kernel ='rbf', random_state = 0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```
