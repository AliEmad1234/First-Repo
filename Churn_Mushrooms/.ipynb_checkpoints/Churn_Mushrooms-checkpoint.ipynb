```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

# Churn

```python
def getPolynomial(data, degree=1):
    if degree ==1:
        return data
    from sklearn.preprocessing import PolynomialFeatures
    poly = PolynomialFeatures(degree= degree)
    poly_features = poly.fit_transform(data)
    return poly_features
```

```python
train = pd.read_csv('Churns/TRAIN.csv')
```

```python
train = train.drop('Customer ID' , 1)
```

```python
# solve the categorical data
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

train = train.dropna()
train.iloc[:,10] = encoder.fit_transform(train.iloc[:,10])
train.iloc[:,11] = encoder.fit_transform(train.iloc[:,11])
train.corr()
```

```python
#3,4,5,8,9
training_data =  train.iloc[:, [0,1,2,3,4,6,8,10,11]].values
goal_data     =  train.iloc[:, -1].values



from sklearn.cross_validation import train_test_split
X_train_1, X_test_1, Y_train_1, Y_test_1 = train_test_split(training_data, goal_data, test_size=0.2, random_state = 0)
```

```python
degree = 1
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(random_state = 0)
logistic.fit(X_train_1, Y_train_1)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test_1, logistic.predict(X_test_1))
pd.DataFrame(cm)
```

```python
logistic.score(getPolynomial(X_test_1,degree), Y_test_1)
```

# Mushrooms

```python
mushrooms = pd.read_csv('Mushrooms/mushrooms.csv')
mushrooms.head()
```

```python

# solve the categorical data
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in range(0,23):
    mushrooms.iloc[:,i] = encoder.fit_transform(mushrooms.iloc[:,i])
mushrooms.head()
```

```python
mushrooms.corr()
```

```python
# 1,2,6,8,17,20,21,22

training_data2 =  mushrooms.iloc[:, [2,3,4,5,6,7,13]].values
goal_data2     =  mushrooms.iloc[:, 0].values


from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(training_data2, goal_data2, test_size=0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
logistic2 = LogisticRegression(random_state = 0)
logistic2.fit(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm2 = confusion_matrix(Y_test, logistic2.predict(X_test))
pd.DataFrame(cm2)
```

```python
logistic2.score(getPolynomial(X_test,degree), Y_test)
```
