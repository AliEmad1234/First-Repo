# titanic

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
taitanic_train = pd.read_csv('taitanic/train.csv');
```

# how many from each gender survived or dead

```python
survivedmale = taitanic_train.loc[(taitanic_train.Survived == 1) & (taitanic_train.Sex == 'male')]
unsurvivedmale = taitanic_train.loc[(taitanic_train.Survived == 0) & (taitanic_train.Sex == 'male')]
```

```python
survivedfemale = taitanic_train.loc[(taitanic_train.Survived == 1) & (taitanic_train.Sex == 'female')]
unsurvivedfemale = taitanic_train.loc[(taitanic_train.Survived == 0) & (taitanic_train.Sex == 'female')]
```

```python
surv_unsurv = np.array([survivedfemale.count()[0] , unsurvivedfemale.count()[0] , survivedmale.count()[0] , unsurvivedmale.count()[0]])
surv_unsurv_title = ['SurvivedFemale' , 'UnsurvivedFemale' , 'SurvivedMale' , 'UnsurvivedMale']
surv_new = pd.DataFrame({'surv_unsurv':surv_unsurv , 'surv_unsurv_title':surv_unsurv_title})
fig, ax = plt.subplots(figsize=(10, 5))
pkmn_type_colors = ["#7cfd02", "#ff0000", "#7cfd02", "#ff0000"]
sns.barplot(ax = ax , x='surv_unsurv_title', y= 'surv_unsurv',  palette=pkmn_type_colors , data= surv_new)
plt.xticks(rotation=-45);
```

# The age of the survived and unsurvived people

```python
survivedfemale_age = survivedfemale.mean()[3]
survivedmale_age = survivedmale.mean()[3]
unsurvivedfemale_age = unsurvivedfemale.mean()[3]
unsurvivedmale_age = unsurvivedmale.mean()[3]
```

```python
surv_unsurv = np.array([survivedfemale_age , unsurvivedfemale_age , survivedmale_age , unsurvivedmale_age])
surv_unsurv_title = ['SurvivedFemaleAge' , 'UnsurvivedFemaleAge' , 'SurvivedMaleAge' , 'UnsurvivedMaleAge']
surv_new_age = pd.DataFrame({'surv_unsurvAge':surv_unsurv , 'surv_unsurv_titleAge':surv_unsurv_title})
fig, ax = plt.subplots(figsize=(10, 5))
pkmn_type_colors = ["#cc0000", "#e62e00", "#ff0000", "#800000"]
sns.barplot(ax = ax , x='surv_unsurv_titleAge', y= 'surv_unsurvAge',  palette=pkmn_type_colors , data= surv_new_age)
plt.xticks(rotation=-45);
```

```python
surv_age = taitanic_train.loc[taitanic_train.Survived == 1]
surv_age = surv_age.groupby('Age')['PassengerId'].count()
surv_age.sort_values()
# the overall mean age of the survived people was 24 years old 
```

```python
unsurv_age = taitanic_train.loc[taitanic_train.Survived == 0]
unsurv_age = unsurv_age.groupby('Age')['PassengerId'].count()
unsurv_age.sort_values()
# the overall mean age of the unservived people was 21 years old
```

# How many people in every class

```python
Pclass1_count = taitanic_train.loc[(taitanic_train.Pclass == 1)].count()[0] 
Pclass2_count = taitanic_train.loc[(taitanic_train.Pclass == 2)].count()[0] 
Pclass3_count = taitanic_train.loc[(taitanic_train.Pclass == 3)].count()[0]
```

```python
labels = ['class 1' , 'class 2' , 'class 3']
sizes = np.array([Pclass1_count, Pclass2_count, Pclass3_count])
explode = np.zeros(len(sizes))
explode[2] = 0.07

plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.2f%%', shadow=True, radius=1.7)
plt.axis('equal')
 
plt.show()
```

# Survived Classes 

```python
Pclass1_surv = taitanic_train.loc[(taitanic_train.Pclass == 1) & (taitanic_train.Survived == 1)].count()[0] 
Pclass2_surv = taitanic_train.loc[(taitanic_train.Pclass == 2) & (taitanic_train.Survived == 1)].count()[0] 
Pclass3_surv = taitanic_train.loc[(taitanic_train.Pclass == 3) & (taitanic_train.Survived == 1)].count()[0] 
```

```python
labels = ['class 1' , 'class 2' , 'class 3']
sizes = np.array([Pclass1_surv, Pclass2_surv, Pclass3_surv])
explode = np.zeros(len(sizes))
explode[0] = 0.07

plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.2f%%', shadow=True, radius=1.7)
plt.axis('equal')
plt.show()
```

# Unsurvived Classes 

```python
Pclass1_unsurv = taitanic_train.loc[(taitanic_train.Pclass == 1) & (taitanic_train.Survived == 0)].count()[0] 
Pclass2_unsurv = taitanic_train.loc[(taitanic_train.Pclass == 2) & (taitanic_train.Survived == 0)].count()[0] 
Pclass3_unsurv = taitanic_train.loc[(taitanic_train.Pclass == 3) & (taitanic_train.Survived == 0)].count()[0] 
```

```python
labels = ['class 1' , 'class 2' , 'class 3']
sizes = np.array([Pclass1_unsurv, Pclass2_unsurv, Pclass3_unsurv])
explode = np.zeros(len(sizes))
explode[2] = 0.07

plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.2f%%', shadow=True, radius=1.7)
plt.axis('equal')
 
plt.show()
```
