```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

# Wine Quality

# Can your predict the wine quality for a factory ? (for educational purpose only)

given [this dataset](https://www.kaggle.com/rajyellow46/wine-quality) we want to predict the class of the wine.

```python
wine = pd.read_csv('winequalityN/winequalityN.csv')
wine = wine.dropna()
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
wine.iloc[:,0] = encoder.fit_transform(wine.iloc[:,0])
wine.corr()
# 0,3,6,7,9,10,11       ==  61% B
# 0,1,2,3,4,6,7,10,11   ==  61% 
```

```python
features = wine.iloc[:,[0,1,2,3,4,6,7,10,11]].values
goal = wine.iloc[:,-1].values

# import the oneHotEncoder class
from sklearn.preprocessing import OneHotEncoder
oneHotEncoder = OneHotEncoder(categorical_features=[0])
features = oneHotEncoder.fit_transform(features).toarray()


# standrize the values
#from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#features[:,[1,2,3,4,5,6,8]] = scaler.fit_transform(features[:,[1,2,3,4,5,6,8]])






# split into train and test
from sklearn.model_selection import train_test_split
train_set, test_set, goal_train, goal_test = train_test_split(features,goal,test_size =0.20,random_state=0)
```

# Using Linear SVM

```python
from sklearn.svm import LinearSVC
classifier = LinearSVC(random_state = 0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
for K in [1, 3, 5, 10, 20, 100, 150, len(test_set)]:
    classifier = KNeighborsClassifier(n_neighbors=K)
    classifier.fit(train_set, goal_train)
    print("model accuracy : {:.4f}".format(classifier.score(test_set, goal_test)))

    cm = confusion_matrix(goal_test, classifier.predict(test_set))
    print('K = %d\n-------------'%K)
    print(pd.DataFrame(cm))
    print('-------------\n\n')
```

# Using Logistic

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(random_state = 0)
logistic.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, logistic.predict(test_set))
pd.DataFrame(cm)

logistic.score(test_set, goal_test)
```

# Using Naive Bays Gaussian

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state=0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using Gaussian SVM

```python
from sklearn.svm import SVC
classifier = SVC(kernel ='rbf', random_state = 0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using Poly SVM

```python
classifier = SVC(kernel='poly', random_state=0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Breast Cancer

```python
BCancer = pd.read_csv('breast-cancer/breast-cancer.csv')
BCancer = BCancer.dropna()

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in range(0,10):
    BCancer.iloc[:,i] = encoder.fit_transform(BCancer.iloc[:,i])

BCancer.head()
```

```python
features = BCancer.iloc[:,[0,4,5,6]].values
goal = BCancer.iloc[:,-1].values

# import the oneHotEncoder class
from sklearn.preprocessing import OneHotEncoder
oneHotEncoder = OneHotEncoder(categorical_features=[0,1])
features = oneHotEncoder.fit_transform(features).toarray()


# split into train and test
from sklearn.model_selection import train_test_split
train_set, test_set, goal_train, goal_test = train_test_split(features,goal,test_size =0.20,random_state=0)
```

# Using Linear SVM



```python
from sklearn.svm import LinearSVC
classifier = LinearSVC(random_state = 0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
for K in [1, 3, 5, 10, 20, 100, 150, len(test_set)]:
    classifier = KNeighborsClassifier(n_neighbors=K)
    classifier.fit(train_set, goal_train)
    print("model accuracy : {:.4f}".format(classifier.score(test_set, goal_test)))

    cm = confusion_matrix(goal_test, classifier.predict(test_set))
    print('K = %d\n-------------'%K)
    print(pd.DataFrame(cm))
    print('-------------\n\n')
```

# Using Logistic



```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(random_state = 0)
logistic.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, logistic.predict(test_set))
pd.DataFrame(cm)

logistic.score(test_set, goal_test)
```

# Using Naive Bays Gaussian



```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using Decision Tree



```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state=0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

# Using Gaussian SVM



```python
from sklearn.svm import SVC
classifier = SVC(kernel ='rbf', random_state = 0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```

```python
classifier = SVC(kernel='poly', random_state=0)
classifier.fit(train_set, goal_train)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
print(classifier.score(test_set, goal_test))
pd.DataFrame(cm)
```
