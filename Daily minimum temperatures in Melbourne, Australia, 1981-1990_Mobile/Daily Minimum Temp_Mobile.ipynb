```python
import pandas as pd
```

```python
df = pd.read_excel('daily-minimum-temperatures-in-me.xlsx')
dm = df.copy()
dd = df.copy()
```

## Convert every date to datetime format in every dataframe 

```python
df['Date'] = pd.to_datetime(df.Date)
dm['Date'] = pd.to_datetime(dm.Date)
dd['Date'] = pd.to_datetime(dd.Date)
```

## Set the index is the date in every dataframe

```python
df.set_index(df.Date , inplace=True)
dm.set_index(dm.Date , inplace=True)
dd.set_index(dd.Date , inplace=True)
```

```python
# resample df dataframe by the year
df = df.resample('Y').mean()
# resample df dataframe by the month
dm = dm.resample('M').mean()
# resample df dataframe by the day
dd = dd.resample('D').mean()
```

## Plot the histogram of every dataframe( with some info. )

```python
%matplotlib inline
df['Daily minimum temperatures in Melbourne, Australia, 1981-1990'].plot(kind="hist" , title = 'Yearly' )
# that means that the temp 10.8 c was the most frequently happened and after it is 11.2 and the rest are the same
```

```python
dm['Daily minimum temperatures in Melbourne, Australia, 1981-1990'].plot(kind="hist" , title = 'monthely' )
# this means that the most temp that frequently happened was from 13-14 celsius, after it was from 7-9 celsius, after that 
# was from 14-15.5 celsius 
```

```python
dm['Daily minimum temperatures in Melbourne, Australia, 1981-1990'].plot(kind="hist" , title = 'daily' )
# this means that the temp that most frequently happened in the days was from 12.5-13.5 celsius, after it was 
# from 7-8.5 celsius, then from 14-15 celsius 
```

## Some observation about the results

```python
df.sort_values(by='Daily minimum temperatures in Melbourne, Australia, 1981-1990' , ascending= True)
# this show that the minimum temp happened was in year 1984
# and the highest temp from these low temps was in year 1988
# all these opservations according to the mean of the data along every year
# so the coldest year is 1984
```

```python
dm.sort_values(by='Daily minimum temperatures in Melbourne, Australia, 1981-1990' , ascending= True)
# this show that the minimum temp happened was in month 7 at year 1982
# and the highest temp from these low temps was in month 1 at year 1981
# all these opservations according to the mean of the data along every month
# so the coldest month is 7 at year 1982
```

```python
dd.sort_values(by='Daily minimum temperatures in Melbourne, Australia, 1981-1990' , ascending= True)
# this show that the minimum temp happened was in day 21 at month 7 at year 1982
# and the highest temp from these low temps was in day 31 at month 12 at year 1988
# all these opservations according to the mean of the data along every month
# so the coldest month is 7 at year 1982
```

```python
import numpy as np
#AppleStore.csv
#appleStore_description.csv
applestore = pd.read_csv('AppleStore.csv')
applestore_description = pd.read_csv('appleStore_description.csv')
```

```python
applestore_description.set_index(applestore_description.id , inplace=True);
applestore.set_index(applestore.id , inplace=True);
```

```python
applestore_description['id'] = pd.to_numeric(applestore_description.id)
applestore['id'] = pd.to_numeric(applestore.id)
```

```python
ids = np.array(applestore['id'])
```

```python
#applestore_description.loc[applestore_description.id == ids]
```

```python
applestore.loc[applestore.prime_genre == 'Games'].loc[applestore.user_rating >= 4.0];
# games that have rate from 4 up to 5 is about 2868 game
```

```python
applestore.loc[applestore.prime_genre == 'Games'].loc[applestore.user_rating == applestore.user_rating.min()];
# games with least rate are 462 games and most of it is chinease or japanese
```

```python
userrate = applestore.groupby('prime_genre')['user_rating'].sum()
userrate = userrate.sort_values(ascending=False)
# that means that the best rate is for games applications
```

```python
userrate
```

```python
#applestore.loc[applestore.rating_count_ver == applestore.rating_count_ver.max()]
applestore.loc[applestore.prime_genre == 'Games'].loc[applestore.user_rating_ver == 5].loc[applestore.rating_count_ver == applestore.rating_count_ver.max()]
# this means that the best game according to the rating of the user and the rating count version is infinity Blade
```

```python
applestore.sort_values('rating_count_tot' , ascending=False)
# this means that according to the rating count total that the Facebook is the most app downloaded, then the instagranm
# and  a lot of these apps is free, that means that people download the free apps more than the other apps
# and the least apps, the most is Japanese or chinease apps
```
