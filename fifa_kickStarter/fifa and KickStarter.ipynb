# where should we be searching for the next ronaldinho ?

```python
nextronaldyo = fifa.loc[(fifa.Overall >= 85)]
nextronaldyo = nextronaldyo.loc[(nextronaldyo['Preferred Positions'].str.contains('ST'))]
nationals = nextronaldyo['Nationality']
nationals = sorted(nationals.unique())

labels = nationals
sizes = np.array(nextronaldyo.groupby('Nationality').count()['ST'])
explode = np.zeros(len(sizes))
explode[0] = 0.1
explode[4] = 0.1

plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.2f%%', shadow=True, radius=1.7)
 
plt.show()


# so in Argentina and France we can find the next ronaldinho
```

```python
kickstarter = pd.read_csv('kickstarter\ks-projects-201801.csv');
```

```python
kickstarter
```

```python
success = kickstarter.loc[kickstarter.state == 'successful']
```

```python
success.groupby('main_category')['state'].count()
# as the most succesful category is the music so we should invest in it
```

```python
mainstates = pd.DataFrame({'country':sorted(kickstarter['main_category'].unique()) , 'state':np.array(success.groupby('main_category')['state'].count())})
fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(ax = ax , x='country', y= 'state', data= mainstates)
plt.xticks(rotation=-45);
```

```python
success.groupby('country')['state'].count()
# so US is the best country that we can launch our offline office
```

```python
countrystates = pd.DataFrame({'country':sorted(kickstarter['country'].unique()) , 'state':np.array(success.groupby('country')['state'].count())})
fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(ax = ax , x='country', y= 'state', data= countrystates)
plt.xticks(rotation=-45);
```

```python
info = kickstarter.groupby('state').count()
```

```python
labels = sorted(kickstarter['state'].unique())
sizes = np.array(info['ID'])
explode = np.zeros(len(sizes))
explode[0] = 0.1


plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.2f%%', shadow=True, radius=2)
 
plt.show()
# that means that it failed by 52.22% so this platform is not good
```
